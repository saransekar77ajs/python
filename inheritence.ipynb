```python
# Create a base class with 2 instance variable n,p and then create a sub class to calculate the n power p and return the result

class A:
  def __init__(self,n,p):
    self.n = n
    self.p = p
    
class B(A):
  def calc_power(self):
    return self.n ** self.p

obj = B(2,3)
obj.calc_power()    
```

```python
#Create a super class with employee details with parameters of a method and display in another method
# sub class employee details call allmethod of super class in this method
class Employee:
  def __init__(self, emp_id,emp_name,emp_email,emp_salary):
    self.empId = emp_id
    self.empName = emp_name
    self.empEmail = emp_email
    self.empSalary = emp_salary
    
  def display_info(self):
    print("**********Employee Information**********")
    print(f"Employee Id: {self.empId}")
    print(f"Employee Name: {self.empName}")
    print(f"Employee Email: {self.empEmail}")
    print(f"Employee Salary: {self.empSalary}")
  
class EmployeeInfo(Employee):
  def __init__(self, emp_id, emp_name, emp_email, emp_salary):
    super().__init__(emp_id, emp_name, emp_email, emp_salary)
  def show_details(self):
    self.display_info()  
  
empl1 = EmployeeInfo("emp001","John Doe","johndoe@test.com",175000)  
empl2 = EmployeeInfo("emp002","Sarah Lee","sarahlee@test.com",162000)
empl1.show_details()
empl2.show_details()
    
```

```python
#Create multi level inhertience for Students and calculate and display the results

class Student:
  def __init__(self,id,name):
    self.student_id =id
    self.student_name = name
    
class Marks(Student):
  def __init__(self,id,name,english,maths,science,social_studies,physics):
    super().__init__(id,name)
    self.english = english
    self.maths = maths
    self.science = science
    self.social_studies = social_studies
    self.physics = physics
    
class Result(Marks):
  
  def display_marks(self):    
        print("**********Student Result**********")
        print(f"Student ID: {self.student_id}")
        print(f"Student Name: {self.student_name}")
        print(f"English: {self.english}")
        print(f"Maths: {self.maths}")
        print(f"Science: {self.science}")
        print(f"Social Studies: {self.social_studies}")
        print(f"Physics: {self.physics}")
        
  def calc_total(self):
    return self.english + self.maths + self.science + self.social_studies + self.physics
  
  def cal_average(self):
    average = self.calc_total() / 5
    return average
    
  def grade(self):
    average = self.cal_average()
    failed_subjects = [subj for subj, score in {
        "English": self.english, 
        "Maths": self.maths, 
        "Science": self.science, 
        "Social Studies": self.social_studies, 
        "Physics": self.physics
    }.items() if score <= 45]

    if not failed_subjects:
        if average > 90:
            print(f"Your Average: {average}, 'Grade A'")
        elif average > 80:
            print(f"Your Average: {average}, 'Grade B'")
        elif average > 70:
            print(f"Your Average: {average}, 'Grade C'")
        elif average > 60:
            print(f"Your Average: {average}, 'Grade D'")
        else:
            print(f"Your Average: {average}, 'Grade E'")
    else:
        print(f"You failed in {', '.join(failed_subjects)}. Better luck next time.")  
 

stu1 = Result("001", "John", 89, 82, 95, 93, 73)

print(f"Results for Student: {stu1.student_name}")
print(f"Total Marks: {stu1.calc_total()}")  
print(f"Average Marks: {stu1.cal_average()}") 
stu1.grade() 
print() 

stu2 = Result("002", "Sarah", 89, 32, 45, 93, 73)
print(f"Results for Student: {stu2.student_name}")
print(f"Total Marks: {stu2.calc_total()}")  
print(f"Average Marks: {stu2.cal_average()}") 
stu2.grade() 
print()  


stu3 = Result("003", "Mike", 46, 46, 95, 93, 73)
print(f"Results for Student: {stu3.student_name}")
print(f"Total Marks: {stu3.calc_total()}")  
print(f"Average Marks: {stu3.cal_average()}") 
stu3.grade()  
```

```python
#Heirarichal inheritance
class Data:
  def __init__(self,x,y):
    self.x = x
    self.y =y
  def display(self):
    print(f"Value of x is {self.x}")  
    print(f"Value of y is {self.y}") 
    
class addition(Data):
  def calculate(self):
    return self.x+self.y
  
class subtraction(Data):
  def calculate(self):
    return self.x-self.y
  
  
class multiplication(Data):
  def calculate(self):
    return self.x*self.y
  
class division(Data):
  def calculate(self):
    try:
      return self.x / self.y
    except ZeroDivisionError as e:
      print(f"You cannot divide by zero: {e}") 
      
      
add = addition(10,5) 
add.display()
print("Addition: ",add.calculate()) 

sub = subtraction(10,5) 
sub.display()
print("Subtraction: ",sub.calculate())       

mul = multiplication(10,5) 
mul.display()
print("Multiplication: ",mul.calculate()) 

divi = division(10,5) 
divi.display()
print("Division: ",divi.calculate())        
            
```

```python
class FirstYear:
    def __init__(self):
        self.first_year_marks = {}

    def read_first_year_marks(self):
        print("Enter First Year Marks:")
        self.first_year_marks = {
            "Telugu1": int(input("Telugu: ")),
            "English1": int(input("English: ")),
            "Math1": int(input("Math: ")),
            "Physics1": int(input("Physics: ")),
            "Chemistry1": int(input("Chemistry: "))
        }

    def display_first_year_marks(self):
        print("\nFirst Year Marks:")
        for subject, mark in self.first_year_marks.items():
            print(f"{subject}: {mark}")

class SecondYear:
    def __init__(self):
        self.second_year_marks = {}

    def read_second_year_marks(self):
        print("Enter Second Year Marks:")
        self.second_year_marks = {
            "Telugu2": int(input("Telugu: ")),
            "English2": int(input("English: ")),
            "Math2": int(input("Math: ")),
            "Physics2": int(input("Physics: ")),
            "Chemistry2": int(input("Chemistry: "))
        }

    def display_second_year_marks(self):
        print("\nSecond Year Marks:")
        for subject, mark in self.second_year_marks.items():
            print(f"{subject}: {mark}")

class Result(FirstYear, SecondYear):
    def __init__(self, student_id, name):
        super().__init__()
        self.student_id = student_id
        self.student_name = name
        self.total_marks = 0
        self.percentage = 0
        self.grade = "Fail"
        self.failed_subjects = []

    def calculate_result(self):
        all_marks = self.first_year_marks.copy()
        all_marks.update(self.second_year_marks)
        
        self.failed_subjects = [subject for subject, mark in all_marks.items() if mark < 35]
        
        if not self.failed_subjects:  # Ensuring all subjects are passed
            self.total_marks = sum(all_marks.values())
            self.percentage = (self.total_marks / (10 * 100)) * 100  # Since 10 subjects, 100 marks each
            self.grade = self.assign_grade()

    def assign_grade(self):
        if self.percentage >= 90:
            return "A+"
        elif self.percentage >= 80:
            return "A"
        elif self.percentage >= 70:
            return "B+"
        elif self.percentage >= 60:
            return "B"
        elif self.percentage >= 50:
            return "C"
        elif self.percentage >= 35:
            return "D"
        else:
            return "Fail"

    def display_result(self):
        print("\n-------- Student Academic Report --------")
        print(f"Student ID: {self.student_id}")
        print(f"Student Name: {self.student_name}")
        self.display_first_year_marks()      
        self.display_second_year_marks()

        if self.grade != "Fail":
            print("\nResult: PASS")
            print(f"Total Marks: {self.total_marks}")
            print(f"Percentage: {self.percentage:.2f}%")
            print(f"Grade: {self.grade}")
        else:
            print("\nResult: FAIL")
            print("Failed Subjects:", ", ".join(self.failed_subjects))

# Example usage
student = Result("001", "John")
student.read_first_year_marks()
student.read_second_year_marks()
student.calculate_result()
student.display_result()
```

```python
# Hybrid inheritance
class College:
  def __init__(self,college_id,college_name,building_no, street,city,state,country,zip, student_id,student_name):
    self.college_id = college_id
    self.college_name= college_name
    self.college_address = {"building_no":building_no,
      "street":street, "city":city,"state":state,"country":country,"zip":zip}
    self.student_id = student_id
    self.student_name = student_name
    
  def display_info(self):    
    print(f"College Name: {self.college_name}")
    print(f"Student Id: {self.student_id}")
    print(f"Student Name: {self.student_name}")
    print(f"College Address: {self.college_address['city']}, {self.college_address['state']}")
    
class FirstYear(College):
    def __init__(self, college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name):        
        super().__init__(college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name)
        self.first_year_marks = {}


    def read_first_year_marks(self):
        print("Enter First Year Marks:")
        self.first_year_marks = {
            "Telugu1": int(input("Telugu: ")),
            "English1": int(input("English: ")),
            "Math1": int(input("Math: ")),
            "Physics1": int(input("Physics: ")),
            "Chemistry1": int(input("Chemistry: "))
        }

    def display_first_year_marks(self):
        print("\nFirst Year Marks:")
        for subject, mark in self.first_year_marks.items():
            print(f"{subject}: {mark}")
            
class SecondYear(College):
  def __init__(self, college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name):
        super().__init__(college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name)
        self.second_year_marks = {}


  def read_second_year_marks(self):
        print("Enter Second Year Marks:")
        self.second_year_marks = {
            "Telugu2": int(input("Telugu: ")),
            "English2": int(input("English: ")),
            "Math2": int(input("Math: ")),
            "Physics2": int(input("Physics: ")),
            "Chemistry2": int(input("Chemistry: "))
        }

  def display_second_year_marks(self):
        print("\nSecond Year Marks:")
        for subject, mark in self.second_year_marks.items():
            print(f"{subject}: {mark}")
              
class Result(FirstYear, SecondYear):
    def __init__(self, college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name):
        super().__init__(college_id, college_name, building_no, street, city, state, country, zip, student_id, student_name)       
        self.total_marks = 0
        self.percentage = 0
        self.grade = "Fail"
        self.failed_subjects = []

    def calculate_result(self):
        all_marks = self.first_year_marks.copy()
        all_marks.update(self.second_year_marks)
        
        self.failed_subjects = [subject for subject, mark in all_marks.items() if mark < 35]
        
        if not self.failed_subjects:  # Ensuring all subjects are passed
            self.total_marks = sum(all_marks.values())
            self.percentage = (self.total_marks / (10 * 100)) * 100  # Since 10 subjects, 100 marks each
            self.grade = self.assign_grade()

    def assign_grade(self):
        if self.percentage >= 90:
            return "A+"
        elif self.percentage >= 80:
            return "A"
        elif self.percentage >= 70:
            return "B+"
        elif self.percentage >= 60:
            return "B"
        elif self.percentage >= 50:
            return "C"
        elif self.percentage >= 35:
            return "D"
        else:
            return "Fail"

    def display_result(self):
        print("\n-------- Student Academic Report --------")
        # print(f"Student ID: {self.student_id}")
        # print(f"Student Name: {self.student_name}")
        self.display_info()
        self.display_first_year_marks()      
        self.display_second_year_marks()

        if self.grade != "Fail":
            print("\nResult: PASS")
            print(f"Total Marks: {self.total_marks}")
            print(f"Percentage: {self.percentage:.2f}%")
            print(f"Grade: {self.grade}")
        else:
            print("\nResult: FAIL")
            print("Failed Subjects:", ", ".join(self.failed_subjects))              

student = Result("CUNY234567","City University Of New York",1234,"LongIsland Avenue","Queens","New York","USA",11372,"stu001","John")
student.read_first_year_marks()
student.read_second_year_marks()
student.calculate_result()
student.display_result()    
    
```
